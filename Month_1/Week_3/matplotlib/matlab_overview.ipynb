```python
x= [1,2,3]
y= [2,4,6]
# size of the graph
plt.figure(figsize=(12,6),dpi=300)
x2=np.arange(0,3)
# ploting the graph with legends
plt.plot(x,y,label='2x',color='g',linewidth=2,marker='.',markersize=8,linestyle='--')
#used to label the line and title
plt.title('Our first Graph!',fontdict={'fontname':'FreeSans','fontsize':24})
plt.xlabel("x axis")
plt.ylabel("y axis")
# overwrite the x label strucuture
plt.xticks([0,1,2,3])
plt.plot(x,x2,label='1x')
plt.yticks([0,2,4,6])
plt.legend()
# saves the graph 
plt.savefig('mygraph.png',dpi=200)
plt.show()
```

## barChart

```python
bars = plt.bar(x,y)
# pattersn
bars[0].set_hatch('o')
bars[1].set_hatch('*')
bars[2].set_hatch('+')
plt.show()
```
