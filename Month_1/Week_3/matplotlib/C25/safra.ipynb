```python
import matplotlib.pyplot as plt 
%matplotlib inline 
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2+x
y
```

```python
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Title")
plt.xlim(0,7)
plt.ylim(2,9)
plt.show()
```

```python
a  = np.linspace(0,10,11) 
b = a**4 
```

```python
x = np.arange(0,10) 
y = 2*x
```

```python
fig = plt.figure()
fig
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(12,7),dpi=120)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes1.plot(a,b)
axes1.set(xlabel='x axis',ylabel='y axis',title="a and b representation")
axes2.plot(x,y)
axes2.set(xlabel='x axis',ylabel='y axis',title='x and y representation')
plt.show()
```
