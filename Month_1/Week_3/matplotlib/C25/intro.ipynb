## What does %matplotlib inline do?

- Displays plots directly in the notebook – When you use this command, Matplotlib plots will appear as static images embedded in the notebook output cells.
  
- No separate window – Without this command, some backends (like Qt or Tk) might open plots in an external window, but inline keeps them inside the notebook.

- Static images – The plots are rendered as PNG (by default) and displayed below the code cell.

```python
# %matplotlib notebook 
```

- Provides interactive plots (zoom, pan) within the notebook.
