## These two lines of code are fundamental for creating a figure and an axes object in Matplotlib, which are essential for plotting.
1. `fig = plt.figure()`

- Purpose: Creates a new figure (a blank canvas/window where plots will be drawn).

- What it does:
        - Initializes an empty figure container.
        - You can customize the figure size, DPI (dots per inch), and background color.
- Example with customization:

```python
    fig = plt.figure(figsize=(8, 6), dpi=100, facecolor='lightgray')
```
- figsize=(width, height) → Controls dimensions in inches.
- dpi → Adjusts resolution (default is usually 100).
- facecolor → Sets the background color.
2.`ax = plt.axes()`

- Purpose: Creates an axes object (a single subplot where data is plotted).
- What it does:
        - Adds a set of x-y axes to the figure.
        - By default, it creates one subplot (1 row × 1 column).

- Example with customization:
```python

ax = plt.axes(xlabel='X-axis', ylabel='Y-axis', title='My Plot')
```
- You can later modify axes properties like labels, ticks, and gridlines.

````python
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x),color='blue',label="sin(x)")
ax.plot(x,np.cos(x),color='green',label="cos(x)")
plt.legend()
plt.show()
````

````python
ax = plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-2,2),
       xlabel='x',ylabel='sin(x)',
       title='A Simple Plot')
plt.show()
````
