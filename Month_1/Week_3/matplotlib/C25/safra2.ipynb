```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
a = np.linspace(0,10,11) 
b = a**4 
x = np.arange(0,10) 
y = 2*x
```

```python
fig,axes = plt.subplots()
axes.plot(x,y)
axes.set(xlabel='x axis',ylabel='y axis')
axes.plot(a,b)
plt.show()
```
