## 2) How many missing data points do we have?

What percentage of the values in the dataset are missing? Your answer should be a number between 0 and 100. (If 1/4 of the values in the dataset are missing, the answer is 25.)

```python
missing_values_count = permits_data.isnull().sum()
print(f"missing values count {missing_values_count} ")
total_missing_cells = missing_values_count.sum()
total_original_cells = np.prod(permits_data.shape)
print("----------------------\n")
print(f"Total missing cells = {total_missing_cells} \n Total original dataset = {total_original_cells} ")
```
